# Ramdom Forest Regression

## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import data set

```python
df = pd.read_csv('Position_Salaries.csv')
# Independent variables
X = df.iloc[:, 1:-1].values
# Dependent variable
y = df.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
# n_estimator: number of tree
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

## Predicting new result

```python
regressor.predict([[6.5]])
```

## Visualizing the Random Forest Regression model on the whole dataset (Higher resolution)

```python
X_grid = np.arange(min(X.reshape(-1)), max(X.reshape(-1)), 0.1)

plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid.reshape(-1,1)), color='blue')
plt.title('Random forest regression') 
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
